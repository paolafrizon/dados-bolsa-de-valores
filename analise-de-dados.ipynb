```python
import pandas as pd
import plotly
import plotly.offline as py
import plotly.graph_objs as go
```

```python
df = pd.read_csv('export/PETR4.SA.csv')
```

```python
df
```

#### Primeira visualização (simples)

```python
data = [go.Scatter(
x = df.Date,
y = df['Close'])]

py.iplot(data)
```

#### Customizando Faixa de eixos 

```python
Close = go.Scatter(
x=df.Date,
y=df.Close,
name = "PETR4 Close",
line = dict(color = '#e53935'),
opacity = 0.8)

data = [Close]

layout = dict(
title = "Faixa para exemplo de customização",
title_x= 0.5,
xaxis = dict(
range = ['2020-04-07','2021-04-06'])
)

fig = dict(data=data, layout=layout)
py.iplot(fig, filename = "Manually Set Range")
```
